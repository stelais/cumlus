# Resolution
The resolution of image can be affected by aberration and by diffraction. In a case where the lens aperture is very **narrow**, the Point Spread Function (PSF) is dominated by **diffraction**. Therefore, we are going to work with the diffraction scenario, in which the lens' circular aperture is analogous to a 2D version of a single-slit experiment. 

## The Rayleigh Criterion:
$\theta = 1.22 \frac{\lambda}{D}$ 

```python
def angular_resolution(wavelength_, diameter_aperture_):
  resolution__radians_ = 1.22 * wavelength_ / diameter_aperture_
  return angular_resolution
```

If point-like sources are separated by angle smaller than the angular resolution, they will not be resolved as separated sources. 
 

## Pixel 
For Charge-Coupled Device (CCD) image sensors, pixels are the p-doped Metal-Oxide-Semiconductor (MOS) capacitors
* photoactive region
* transmission region (shift register)

Each capacitor accumulates an electric charge proportional to a light intensity, and then each capacitor transfers its content to its neighbor. A charge amplifier converts the charge of the last capacitor into a voltage.
